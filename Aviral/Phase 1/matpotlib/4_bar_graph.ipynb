```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMAZON', 'MICROSOFT', 'META']
revenue= [90,123,98,76]
profit=[80,100,90,70]
```

```python
xpos= np.arange(len(company))
plt.xticks(xpos,company) #replace indexing with namee
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue') #will not take string value
plt.xlabel('company')
plt.ylabel('revenue')
plt.bar(xpos+0.2 ,profit,width=0.4,label='profit')
plt.legend()
#barh : horizontal bar
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,'--r+') #you can specify in variety of symbols and color 
plt.show()
```

```python
plt.plot(x,y,color='red', marker ='D',linestyle=' ')
plt.show()
```

```python
# alpha property
plt.plot(x,y,color='red', marker ='D',alpha=0.2)
plt.show()
```

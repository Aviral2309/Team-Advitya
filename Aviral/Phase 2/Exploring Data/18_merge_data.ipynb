```python
import numpy as np
import pandas as pd
import os
```

```python
table1 = pd.DataFrame({"P_ID" :(1,2,3,4,5,6,7,8),
                       "gender":('m','m','f','f','f','m','m','f'),
                       "height":(71,74,78,75,70,69,72,75)})
table1
```

```python
table2 = pd.DataFrame({"P_ID" :(1,2,3,4,5,6,7,9),
                       "sex":('m','m','f','f','f','m','m','f'),
                       "visit":(1,4,2,5,0,4,2,1),
                       "illness":(0,0,2,3,2,4,2,1)})
table2
```

```python
combined1 = pd.merge(table1,table2, how ='inner', on="P_ID")
combined1
```

```python
# a left join keeps all key values in the first (left) data frame
left_join = pd.merge(table1,table2, how='left', on="P_ID")
left_join  # right for right join
```

```python
# use outer for all rows 
```

```python
table1.rename(columns = {"gender":"sex"}, inplace = True)  # rename gender column
combined2 = pd.merge(table1, table2, how ='outer', on=["P_ID","sex"])
combined2
```

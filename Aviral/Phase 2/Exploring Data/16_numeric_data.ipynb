```python
colmeans = cars.sum()/cars.shape[0]
colmeans
```

```python
centered = cars - colmeans
centered.describe()    #data centere
```

```python
# standard deviation
column_deviation = cars.std(axis =0)
centered_and_scaled = centered/column_deviation
centered_and_scaled.describe()
```
